## Q1. Explain why we have to use the Exception class while creating a Custom Exception.
### Note: Here Exception class refers to the base class for all the exceptions.

#### When creating a custom exception, it is recommended to inherit from the base 'Exception' class.  The Exception class serves as the foundation for all exceptions in the language's exception hierarchy.
#### it can help to improve the clarity, modularity, reusability and maintainability of our code.

```python
import logging
logging.basicConfig(filename = "custom.log", level = logging.INFO)#creating log file to store the data

class vaildage(Exception): #creating custom exception class
    def __init__(self, age):
        self.age = age
```

```python
def vaildateage(age):
    if age < 0:
        raise vaildage("entered age is negative")
    elif age > 200:
        raise vaildage("entered age is very vey high")
    else:
        logging.info("age is vaild")
```

```python
try:
    age = int(input("enter your age: "))
    logging.info(age) # using logging to store input data
    vaildateage(age)
except vaildage as e:
    logging.info("Give vaild age, {} ".format(e))
```

## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

#### 1.FloatingPointError
#### 2.OverflowError
#### 3.ZeroDivisionError
#### These errors are defined in the AithmeticError.

#### ZeroDivisionError: This exception is raised when we trying to divide a number by zero.
 #### Example:

```python
import logging
logging.basicConfig(filename = "ZeroDivision.log", level = logging.INFO)#creating log file to store the data

try:
    a = 10/0
except ZeroDivisionError as e:
    logging.info("Error: zero division error {} ".format(e))
    
```

#### OverflowError: This exception is raised when the result of an arithmetic operation exceeds the maximum representable value.
#### Example:

```python
import math
import logging
logging.basicConfig(filename = "OverFlow.log", level = logging.INFO)#creating log file to store the data

try:
    result = math.exp(1000)
except OverflowError as e:
    logging.info("Error: overflow  error {} ".format(e))
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

#### The LookupError class is a base class for exceptions that occur when a lookup or indexing operation fails. 
####  It serves as a superclass for specific lookup-related exception classes, such as KeyError and IndexError
#### The LookupError class is used to catch and handle exceptions related to unsuccessful lookups or indexing operations, where the specified key or index is not found or is invalid.

### KeyError: This exception is raised when a dictionary key is not found during a lookup operation.

```python
import logging
logging.basicConfig(filename = "key.log", level = logging.INFO)#creating log file to store the data

d = {'a': 1, 'b': 2, 'c': 3}

try:
    key = d['d']
except KeyError as e:
    logging.info("this is key error {} ".format(e)) 
```

### IndexError: This exception is raised when an invalid index is used to access an element in a sequence (e.g., list, tuple, or string).

```python
import logging
logging.basicConfig(filename = "index.log", level = logging.INFO)#creating log file to store the data

l1 = [2,3,4,5]

try:
    a = l1[5]
except IndexError as e:
    logging.info("this is index error {} ".format(e))
```

## Q5. Explain ImportError. What is ModuleNotFoundError?

#### The ImportError exception is raised when an import statement fails to import a module. It can occur due to various reasons, such as the module not being found, the module file being inaccessible, or errors within the module itself.
#### Example:

```python
import logging
logging.basicConfig(filename = "import.log", level = logging.INFO)#creating log file to store the data

try:
    import ram
except ImportError as e:
    logging.info("this is an import error {} ".format(e))
```

#### ModuleNotFoundError: The ModuleNotFoundError is a subclass of ImportError. It specifically indicates that a module cannot be found or imported.
####  Example:



```python
import logging
logging.basicConfig(filename = "Module.log", level = logging.INFO)#creating log file to store the data

try:
    import raj_module
except ModuleNotFoundError as e:
    logging.info("module not found error {} ".format(e))
```
